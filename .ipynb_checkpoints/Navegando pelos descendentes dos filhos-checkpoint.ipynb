```python
from bs4 import BeautifulSoup
import gc
from bs4.element import  NavigableString, Tag
```

```python
with open('arquivo02.html','r') as f:
    soup = BeautifulSoup(f,'lxml')
```

```python
print(len(list(soup.descendants))) 
```

```python
for tag in soup.descendants: # rode entre todos os descendentes 
    if isinstance(tag, NavigableString): # se a tag (elemento) for uma string imprima ela
        print(tag)
    else:
        print(tag.name) # se entrar aqui o elemento não é uma string e por isso tenho que extrair a string
```

```python
for tag in soup.descendants:
    if isinstance(tag, Tag):  # se for uma Tag imprima uma tag
        print(tag)
```

*** Diferença entre Strings e stripped_strings ***

```python
for string in soup.aside.strings: # aside é uma tag do html, desse jeito me retorna varios espaços em branco
    print(repr(string)) # o metodo repr é para gerar representações que o interpretador Python possa entender
```

```python
for string in soup.aside.stripped_strings: # este método retira todos os espaços em branco
    print(repr(string))
```
