```python
from bs4 import BeautifulSoup
from bs4.element import Tag
```

```python
with open('arquivo03.html','r') as f:
    soup = BeautifulSoup(f,'lxml')
```

```python
print(soup.div.parent) # os irmãos de div
```

```python
print(soup.div.next_element.next_element) # ando dois irmãos para frente
```

```python
print(soup.li.previous_element.previous_element) # ando dois irmãos para trás
```

```python
for e in soup.ul.next_elements:
    if isinstance(e,Tag):
        print(e)
```

```python
for e in soup.ul.previous_elements:
    print(repr(e))
```
