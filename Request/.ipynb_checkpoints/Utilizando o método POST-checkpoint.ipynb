```python
import requests
```

```python
url = 'http://www.buscacep.correios.com.br/sistemas/buscacep/resultadoBuscaCepEndereco.cfm?t'
```

```python
payload = {'EndRow':'10','Metodo':'listaLogradouro','relaxation':'13015100',
           'StartRow':'1','TipoConsulta':'relaxation'}
```

```python
response = requests.post(url,data=payload)
```

```python
with open('correios.html','w') as f:
    f.write(response.text)
    
with open('correios.html','r', encoding='utf-8') as g:
    print(g.readlines())
```
