```python
import requests
```

```python
url = 'https://www.youtube.com/watch?'
```

```python
payload = {'search_query' : 'como fazer molho de tomate'}
```

```python
r = requests.get(url, params=payload)
```

```python
print(requests.get(url, params=payload).text) # retorna o texto da pagina da pag que sera assessada
```

```python
print(requests.get(url, params=payload).encoding) # retorna o encoding usado
```

```python
print(requests.get(url, params=payload).url) # retorna a url
```

```python
# salva o conteudo da busca em um arquivo, mas tarde posso usar o Beautifull Soup nele
with open('youtube.html','w') as f:
    f.write(r.text)
```

```python
# verifica se realmente foi salvo
with open('youtube.html','r') as l:
    for line in l.readlines():
        print(line)
```
